```python
import itertools
def func(res):
    return list(set(itertools.chain.from_iterable(res.values)))
```

```python
tagw = TrackArtistsGateway()
agw = ArtistsGateway()
tgw = TracksGateway()
df_t = tgw.fetch_all().drop(columns=['genres'])
df_ta = tagw.fetch_all()
df_a = agw.fetch_all()
df1 = df_ta.merge(df_a, how='left', left_on='artist_id', right_on='id').groupby('track_id')['genres'].apply(func)
df2 = df_t.merge(df1, how='left', left_on='id', right_on='track_id')
df2.head()
```
